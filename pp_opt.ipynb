```python
import pandas as pd
import glob 
import numpy as np
```

```python
results = glob.glob('*studies.csv')
```

```python
avg_results = {}
for result in results:
    res = pd.read_csv(result, index_col=(0, 1))
    params = res.iloc[:, 2:].describe(include='all')
    means = params.loc['mean', :].dropna().to_dict()
    try:
        top = params.loc['top', :].dropna().to_dict()
        params_f = {**means, **top}
    except KeyError:
        params_f = means
    avg_results[result.replace('_studies.csv','')] = params_f
```

```python
pd.DataFrame(avg_results).T
```

```python
ints = ['max_depth', 'n_estimators','min_child_weight', 'min_samples_split']
for model in avg_results:
    for param in avg_results[model]:
        #print(model, param)
        if param in ints:
            avg_results[model][param] = int(np.round(avg_results[model][param], 0))
```
