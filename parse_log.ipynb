```python
import pandas as pd
import numpy as np
```

```python
f = open("log0.out", "r").readlines()
```

```python
models = ['rf', 'gb', 'xgb0', 'xgb1']
splits = [0, 13, 25, 42, 67]
folds = 5

studies = []
for x in splits:
    studies.extend([(y, f, x) for f in range(folds) for y in models])
```

```python
results = {}
k = 0
for i, l in enumerate(f):
    if "A new study created" in l:
        study = studies[k]
        k += 1
        result_text = f[i+1:i+251]
        result_values = [float(x.split('value: ')[1].split(' ')[0]) for x in result_text]
        result_params = [x.split('{')[1].split('}.')[0] for x in result_text]
        bst = np.argmin(result_values)
        results[study] = [result_values[bst], result_params[bst]]
        #results[study] = pd.Series(result).str.split(' ', expand=True)
```
